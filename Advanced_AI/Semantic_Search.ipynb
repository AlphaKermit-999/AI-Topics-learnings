```python
df = pd.read_csv("olympics.csv")
```

```python
df.columns
```

```python
#df['Name']+ ' from ' + df['Country'] + " won a "+df["Medal"] + df["Sport"]
df["combined"] = df['Name']+ ' ' + df['Country'] + " "+df["Medal"] +' '+ df["Sport"]
```

```python
df["combined"]
```

```python
from sentence_transformers import SentenceTransformer
model = SentenceTransformer('sentence-transformers/all-MiniLM-L6-v2') #embedding=encoder
```

```python
embedding = model.encode(df["combined"].tolist(), show_progress_bar=True)
```

# FAISS = ANN(APPROXIMATE NEARET NEIGHBOR)- 32, array

flatl2=accurate, smaller (dot product)

ivfflat: clustering

HNSW:REAL TIME

pq=memory compression

```python
dimension = embedding.shape[1]
```

```python
import torch
torch.cuda.is_available()
```

```python
# !pip install faiss-cpu
```

```python
import faiss
```

```python
import numpy as np
vectors = np.array(embedding).astype("float32")
```

**Eucledian distance**

```python
index_l2 = faiss.IndexFlatL2(dimension)
```

```python
index_l2.add(vectors)
```

```python
faiss.write_index(index_l2,"index_l2.faiss")
```

```python
i1 = faiss.read_index("index_l2.faiss")
```

```python
query = "top football player"
```

```python
query_vector = model.encode([query]).astype('float32')
```

```python
distance, index = i1.search(query_vector, 2)
```

```python
index
```

```python
for i in index[0]:
  print(df.iloc[i])
```

```python
distance
```

# Clustering

```python
index_l2 = faiss.IndexFlatL2(dimension)
index_ivf = faiss.IndexIVFFlat(index_l2, dimension, 5, faiss.METRIC_L2) # calculation is based on eucledian distance, cluster size = 5
```

```python
index_ivf.train(vectors)
```

```python
index_ivf.add(vectors)
```

```python
faiss.write_index(index_ivf,"index_ivf.faiss")
```

HNSW

```python
hnsw = faiss.IndexHNSWFlat(dimension, 32) # number of neighbours = 32
hnsw.add(vectors)
```

```python
query = "top swimming player"
```

```python
query_vector = model.encode([query]).astype('float32')
```

```python
distance, index = index_ivf.search(query_vector,2)
```

```python
index
```

```python
distance
```

```python
for i in index[0]:
  print(df.iloc[i])
```

```python
import chromadb
```

```python
from chromadb.config import Settings

client = chromadb.Client(Settings(
    persist_directory='./chroma_db'
))
```

```python
collection = client.get_or_create_collection(name="my_collections")
```

```python
documents = ['A "temp live API" typically refers to a live weather API that provides real-time temperature data for a specified location. These APIs are used by developers to integrate current weather conditions, including temperature, into their applications, websites, or services.',
             'Beyond current temperature, many APIs offer historical weather data, forecasts (hourly, daily), and sometimes even weather alerts or specialized data like air quality information.',
             'A vector database stores numerical representations of complex, often unstructured, data as multi-dimensional vectors called embeddings. Unlike traditional databases that rely on exact matches, vector databases perform semantic similarity searches, enabling applications to find conceptually related items by analyzing their proximity in the multi-dimensional vector space. They are crucial for modern AI applications, including large language models (LLMs), providing a knowledge base for context-aware search, recommendations, and retrieving relevant information to "ground" AI responses.']
```

```python
ids1 = ['id1', 'id2', 'id3']
```

```python
ids1 = ["id"+str(i) for i in range(1, len(documents)+1)] #uuid
```

```python
from sentence_transformers import SentenceTransformer
model = SentenceTransformer('sentence-transformers/all-MiniLM-L6-v2')
```

```python
embeddings = model.encode(documents).tolist()
```

```python
collection.add(
    documents=documents,
    embeddings=embeddings,
    ids=ids1
)
```

```python
query = "where can i access live television"
```

```python
vq = model.encode([query]).tolist()
```

```python
collection.query(vq, n_results=2)
```
