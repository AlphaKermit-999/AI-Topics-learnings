```python
from transformers import AutoTokenizer, AutoModelForCausalLM
model_name = "Qwen/Qwen3-0.6B"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForCausalLM.from_pretrained(model_name)
import torch
```

```python
prompt = "Artificial intelligence is human intelligence"
```

```python
input1 = tokenizer(prompt, return_tensors="pt")
```

```python
output = model.generate(**input1)
tokenizer.decode(output[0])
```

```python
output = model.generate(**input1, max_new_tokens=100)
tokenizer.decode(output[0])
```

```python
output = model.generate(**input1, max_new_tokens=100, temperature=0.9)
tokenizer.decode(output[0])
```

```python
output = model.generate(**input1, temperature=0.9, do_sample=True)
tokenizer.decode(output[0])
```

```python
output = model.generate(**input1, temperature=0.9, do_sample=True, top_p=0.95)
tokenizer.decode(output[0])
```

```python
output = model.generate(**input1, temperature=0.9, do_sample=True, top_p=0.95, )
tokenizer.decode(output[0])
```
