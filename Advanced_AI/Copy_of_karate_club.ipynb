```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
G = nx.karate_club_graph()
```

```python
print("Graph Information:")
print(f"Number of nodes (club-members): {G.number_of_nodes()}")
print(f"Number of edges (connections): {G.number_of_edges()}\n")
```

```python
print("Node Details:")
for node, attributes in G.nodes(data=True):
    print(f"Node {node}: {attributes}")
```

```python
print("\nEdge Details:")
for edge in G.edges(data=True):
    print(f"Edge {edge[0]} - {edge[1]}: {edge[2]}")
```

```python
print("\nDegree of Each Node:")
for node, degree in G.degree():
    print(f"Node {node} has degree {degree}")
```

```python
plt.figure(figsize=(14, 12))
labels = [1 if G.nodes[i]['club'] == 'Officer' else 0 for i in G.nodes]

pos = nx.spring_layout(G)
nx.draw_networkx(
    G, pos, with_labels=True, node_color=labels, cmap='coolwarm', node_size=800, font_size=10
)

plt.title("Zachary's Karate Club Graph with Faction-Based Node Colors", fontsize=14)
plt.show()
```

```python
print("\nAdditional Graph Properties:")
print(f"Is the graph connected? {'Yes' if nx.is_connected(G) else 'No'}")
print(f"Number of connected components: {nx.number_connected_components(G)}")
print(f"Clustering coefficient: {nx.average_clustering(G):.2f}")
print(f"Graph density: {nx.density(G):.4f}")
```
